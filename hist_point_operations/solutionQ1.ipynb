<p>Power law transformation is represented by the equation:</p>

<math class="math" xmlns="http://www.w3.org/1998/Math/MathML">
    <mi>S</mi>
    <mo>(</mo>
    <mi>r</mi>
    <mo>)</mo>
    <mo>=</mo>
    <mi>C</mi>
    <mo>&#8290;</mo>
    <msup>
        <mi>r</mi>
        <mi>Y</mi>
    </msup>
</math>

</body>
</html>


```python
def PowerLawTransformation(r, Y = 1 ,C = 1, strech_in_range = False, T = 0):
    # Power law transformation function 
    # r image before PowerLawTransformation 
    # Y,C constants 
    # return s image after PowerLawTransformation 
    indices1 = np.where( r >= T)
    indices2 = np.where( r < T )
    none_zero_one_img = False
    if r.max() > 1 :
        none_zero_one_img = True
        r = r/255 # if the dynamic domain i between 0 and and 255, normalize the pixels to the range 0 to 1
      
    if not strech_in_range :
      s = C * np.power(r, Y)
    else:
      s = np.zeros(r.shape)
    
      s[indices2]= r[indices2]
      s[indices1]= C * np.power(r[indices1], Y)
       
    if none_zero_one_img :
        s = np.round(s * 256)
    return s
```

```python
def histEqualization(origin_img):

    
    none_zero_one_img = False
    float_image = origin_img
    if origin_img.max() > 1 :
        none_zero_one_img = True
        float_image = origin_img/255.0 # if the dynamic domain i between 0 and and 255, normalize the pixels to the range 0 to 1 
        
    float_hist, gray_scales = exposure.histogram(float_image) 
    total_pixels=float_hist.sum()
    norm_hist = float_hist / total_pixels


    cdf = np.cumsum(norm_hist)
    cdf_hest , bins = exposure.histogram(cdf*255)
    plt.figure(figsize=(10, 4))
    plt.bar(bins, cdf_hest, width=1)
    plt.title("commulative function")
    plt.xlabel('x')
    plt.ylabel('F(x)')
    
    HE_img = cdf[origin_img]
    if none_zero_one_img :
      HE_img=HE_img*255
    
    return HE_img
    
```

<p>We did not achieve histogram equalization in the original picture. Before histogram equalization, we have approximately <strong>X = 100000</strong> pixels of gray level 0, which is the highest ratio relative to the total number of pixels. This ratio is approximately <strong>(X/182024) * 255</strong>, which results in a range of gray levels around <strong>[135, 140]</strong>. This means that when applying the cumulative function, most of the pixels will be transformed to this range, becoming the minimum gray level to which most pixels will be transformed.</p>

<p>Other columns' ratios do not contribute significantly to the total ratio due to the lack of pixels in specific gray levels higher than zero. Consequently, we observe that the pixels between <strong>0 < gray-level < 100</strong> have been transformed to a range of levels around <strong>[135, 140]</strong> to <strong>&lt; 220</strong>. However, these pixels will still have the same ratio but will appear brighter.</p>

<p>This happens due to the nature of the cumulative function, which first accumulates the total ratios up to a specific gray level. Since the gray level of <strong>0</strong> contributes the most to the cumulative function and all other levels are higher than zero and do not have many pixels relative to the gray level of <strong>0</strong>, they will not contribute significantly to the cumulative function. Therefore, they will also be directly affected by the transformation of the <strong>0</strong> gray level.</p>

```python
def histEqualization(origin_img):
    none_zero_one_img = False
    float_image = origin_img
    if origin_img.max() > 1 :
        none_zero_one_img = True
        float_image = origin_img/255.0 # if the dynamic domain i between 0 and and 255, normalize the pixels to the range 0 to 1 
        
    float_hist, gray_scales = exposure.histogram(float_image) 
    total_pixels=float_hist.sum()
    norm_hist = float_hist / total_pixels


    cdf = np.cumsum(norm_hist)
    cdf_hest , bins = exposure.histogram(cdf*255)
    plt.figure(figsize=(10, 4))
    plt.bar(bins, cdf_hest, width=1)
    plt.title("commulative function")
    plt.xlabel('x')
    plt.ylabel('F(x)')
    
    HE_img = cdf[origin_img]
    if none_zero_one_img :
      HE_img=HE_img*255
    
    return HE_img
    
```
